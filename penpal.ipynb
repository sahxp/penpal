```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def user_with_bio(bio_data,user_bio):
    

    nltk.download('stopwords')
    nltk.download('punkt')

    stop_words = set(stopwords.words("english"))
    stemmer = PorterStemmer()

    def preprocess_document(document):
        # Convert to lowercase
        document = document.lower()

        # Remove punctuation and numbers
        document = re.sub(r'[^\w\s]|\d', '', document)

        # Tokenize the document into words
        words = word_tokenize(document)

        # Remove stop words
        words = [word for word in words if word not in stop_words]

        # Stem the words
        words = [stemmer.stem(word) for word in words]

        return words

#     document = "This is a sample document. It contains numbers and punctuation. 123! 456."
#     preprocessed_document = preprocess_document(document)

    # Preprocess the bios and user bio
    preprocessed_bios = [preprocess_document(bio) for bio in bio_data]
    preprocessed_user_bio = preprocess_document(user_bio)

    # Convert the preprocessed bios and user bio back to strings
    preprocessed_bios = [' '.join(bio) for bio in preprocessed_bios]
    preprocessed_user_bio = ' '.join(preprocessed_user_bio)

    # Use TfidfVectorizer to find the TF-IDF representation of the preprocessed bios and user bio
    vectorizer = TfidfVectorizer()
    tfidf_bios = vectorizer.fit_transform(preprocessed_bios)
    tfidf_user_bio = vectorizer.transform([preprocessed_user_bio])

    # Use cosine similarity to find the similarity between the user bio and the bios
    similarities = cosine_similarity(tfidf_user_bio, tfidf_bios)

    # Find the index of the 5 most similar bios
    top_5 = np.argsort(similarities[0])[-5:][::-1]

    # Print the 5 most similar bios
    return top_5


    #     print("Preprocessed Document:", preprocessed_document)
```

```python
top_5_lst = []
for i in df['Bio']:
    top_5_lst.append(user_with_bio(bio_data,i))
df['top_5']  = top_5_lst
df.head()
```

```python
def jaccard_similarity(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    
    return len(intersection) / len(union)


# similarity = jaccard_similarity(user1_similar_bios, user2_similar_bios)

# print("Jaccard Similarity:", similarity)
```

```python
top_5_recomend = []
a=0
for i in df['top_5']:
    a +=1
    b=0
    recommend_5_lst =[]
    for j in df['top_5']:
        b+=1
        if a != b:
            recommend_5_lst.append(jaccard_similarity(i,j))
    top_5_recomend.append(recommend_5_lst)
df['top_5_recommend'] = top_5_recomend
```
